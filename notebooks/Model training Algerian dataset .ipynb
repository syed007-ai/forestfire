```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
df= pd.read_csv("Alergian_forest_fires_cleaned_dataset.csv")
df.head()
```

```python
df.columns 
```

```python
## drop day, month, year 
df = df.drop(["day","month","year"], axis =1)
df.head()
```

```python
df['Classes'].value_counts()
```

```python
# encoding 
df["Classes"]= np.where(df["Classes"].str.contains("not fire"),0,1)
```

```python
df.head()
```

```python
df.tail()
```

```python
df["Classes"].value_counts()
```

```python
## Independent and dependent features : 
X = df.drop("FWI", axis =1)
y = df["FWI"]
```

```python
X.head()
```

```python
y
```

```python
## Train, Test, Split 
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y , test_size= 0.25, random_state= 42 )
```

```python
X_train.shape, X_test.shape
```

```python
##Feature selection based on correlation :
X_train.corr()
```

# Feature Selection : 

```python
## check for multi-collinearity :
plt.figure(figsize=(12,10))
corr = X_train.corr()
sns.heatmap(corr, annot = True)
```

```python
X_train.corr()
```

```python
X_train.corr()
```

```python
def correlation(dataset, threshold):
    col_corr = set() # ek set variable banaru, so that repeated values nahi aaye.. 
    corr_matrix = dataset.corr() # correlation matrix 
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i,j]) > threshold : #if absolute 
                colname = corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr
```

```python
## threshold -- Domain expertise 
corr_features = correlation(X_train, 0.85 )
## jo jo independent features ka corr 0.85 hain, vo isme aayega..
```

```python
corr_features
```

```python
## drop features when correlation is more than 0.85
X_train.drop(corr_features, axis = 1, inplace = True )
X_test.drop(corr_features, axis = 1, inplace = True )
X_train.shape, X_test.shape
```

#  Feature scaling or Standardization : 

```python
from sklearn.preprocessing import StandardScaler 
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train) # z-score formula applying..
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled
```

# Box plots to understand effect of Standard Scaler :

```python
plt.subplots(figsize = (15,5)) 
plt.subplot(1, 2, 1)
sns.boxplot(data = X_train)
plt.title("X_train before scaling ")
plt.subplot(1, 2, 2)
sns.boxplot(data = X_train_scaled)
plt.title("X_train after Scaling ")
# scaled down data.. 
# -- optimization and convergence karne easy hojata..
```

## # Linear Regression Model : 

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, r2_score 
```

```python
linreg = LinearRegression()
linreg.fit(X_train_scaled, y_train)
y_pred = linreg.predict(X_test_scaled)
mae = mean_absolute_error(y_test, y_pred) #mean absolute error, cuz bhot saare outliers hain..
score = r2_score(y_test, y_pred)
print("Mean absolute error : " , mae)
print("R2 Score : ", score)
```

# Lasso Regression : 

```python
from sklearn.linear_model import Lasso ## feature selection ke liye, jo feature ka imp nahi rehta, usko apne aap hi kam krdega
from sklearn.metrics import mean_absolute_error , r2_score
```

```python
lasso = Lasso() 
lasso.fit(X_train_scaled, y_train)
y_pred = lasso.predict(X_test_scaled)
mae = mean_absolute_error(y_test, y_pred)
score = r2_score(y_test, y_pred)
print("Mean absolute error : " , mae)
print("R2 score : " , score)
```

# Ridge Regression : 

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_absolute_error , r2_score
```

```python
ridge = Ridge()
ridge.fit(X_train_scaled, y_train) # BFL krdgea..
y_pred = ridge.predict(X_test_scaled) 
mae = mean_absolute_error(y_test, y_pred) # error.. 
score = r2_score(y_test, y_pred)
print("Mean absolute error : " , mae)
print("R2 score : " , score)
## hence abhi bhi linear regression hi better perform karra than Ridge and Lasso 
```

## Elastic Net Regression : 

```python
from sklearn.linear_model import ElasticNet
from sklearn.metrics import mean_absolute_error , r2_score
```

```python
elastic= ElasticNet()
elastic.fit(X_train_scaled, y_train) # BFL krdgea..
y_pred = elastic.predict(X_test_scaled) 
mae = mean_absolute_error(y_test, y_pred) # error.. 
score = r2_score(y_test, y_pred)
print("Mean absolute error : " , mae)
print("R2 score : " , score)
```
